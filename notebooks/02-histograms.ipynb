# Histograms

```python
# https://pillow.readthedocs.io/en/latest/handbook/concepts.html
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt

import IPython
import numpy as np
```

```python
width = 256
height = 256
```

```python
def show_images(images, titles=[]):
    num_images = len(images)
    num_titles = len(titles)
    fig, ax = plt.subplots(nrows=2, ncols=num_images, figsize=(15, 10))
    
    for i in range(0, num_images):
        img = np.asarray(Image.open(images[i]).convert('L'))
        ax[0,i].hist(img.flatten(), 128)
        ax[1,i].imshow(img, cmap='gray', vmin=0, vmax=255, interpolation='nearest')
        ax[1,i].axis('off')
        title = titles[i] if i < num_titles else ''
        ax[1,i].set_title(title)
```

```python
images = [
    '../test-images/img-1-intensity.png',
    '../test-images/img-4-intensities.png',
    '../test-images/img-256-intensities-horisontal.png'
]
titles = [
    'Single Intesity', 
    '4 Intensity Values', 
    '256 Intensity Values'
]
show_images(images, titles)
```
